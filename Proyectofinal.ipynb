##1. ¿Qué son los filtros FIR e IIR?
 FIR – Finite Impulse Response

Su respuesta al impulso dura un número finito de muestras.

Siempre son estables.

No requieren retroalimentación (no tienen polos reales).

Pueden hacerse lineales en fase, lo que significa que no distorsionan la forma de la onda, solo su amplitud.

## Ventaja clave en EEG:
  cuando importa no deformar la señal (por ejemplo, al detectar componentes oscilatorios como mu y beta), los FIR son ideales.

IIR – Infinite Impulse Response

Su respuesta al impulso teóricamente nunca termina.

Usan retroalimentación (tienen polos).

No siempre son lineales en fase.

Logran mejor selectividad (filtran más fuerte) con menos coeficientes.

##Ventaja clave en EEG:
 cuando se quiere un filtrado muy “estrecho” y eficiente, por ejemplo, para suprimir ruido de 50/60 Hz o aislar bandas específicas.

 ## 2. ¿Cómo se aplican para resaltar patrones en EEG de MI?

En tareas de Motor Imagery (MI) se busca detectar:

Cambios espectrales: ERD/ERS en bandas mu (8–12 Hz) y beta (13–30 Hz)

Patrones temporales: transiciones durante el inicio de la imaginación motora

Patrones espaciales: diferencias entre canales como C3, Cz, C4
(relacionados con corteza motora)

Los filtros FIR/IIR se utilizan para mejorar cada uno de estos aspectos.

## 3. Uso de FIR e IIR para resaltar patrones en cada dominio
 A. Patrones espectrales (frecuencia)
Objetivo: aislar bandas mu y beta claves en MI.
¿Cómo se hace?

Se aplica un filtro pasa banda:

8–12 Hz (mu)

13–30 Hz (beta)

¿Qué filtro se usa?
Tipo	Por qué
FIR	Muy usados en BCI: no distorsionan fase → las oscilaciones quedan limpias.
IIR	Útiles si se quiere filtrado fuerte con baja latencia en tiempo real.
¿Qué se obtiene?

Clarificación del ERD/ERS:

ERD = disminución de potencia antes/durante MI

ERS = incremento después

 Sin filtrado adecuado, el ERD/ERS se pierde en ruido.

## B. PATRONES TEMPORALES
Objetivo: identificar momentos donde la señal cambia (inicio de MI).
¿Cómo ayudan FIR/IIR?

Se aplica un filtro pasa bajas (LPF) para eliminar ruido de alta frecuencia.

Se usa un filtro notch (FIR o IIR) para eliminar 50/60 Hz.

Se pueden aplicar filtros derivativos FIR para realzar cambios temporales.

¿Qué se resalta?

Transiciones claras en el tiempo:

antes → durante → después de imaginar mover mano/pie

Eliminación de ruido que de otro modo cubriría las variaciones del ERD/ERS.

 En decodificación temporal de MI, los FIR permiten mantener suavidad y no distorsionar la fase.

## C. PATRONES ESPACIALES

Los filtros FIR/IIR NO modifican directamente el espacio, pero son necesarios para:

## 1. Preparar la señal antes de métodos espaciales:

CSP (Common Spatial Patterns)

LDA/CCA

CAR (Common Average Reference)

Si no se filtra correctamente:

el ruido afecta las matrices de covarianza

los patrones espaciales quedan contaminados

## 2. Filtros espaciales FIR :

En BCI se suelen aplicar filtros en cada canal (FIR/IIR) + combinación espacial:

Ejemplo:

Filtrar C3, Cz, C4 en banda mu/beta

Luego calcular CSP

Y así resaltar diferencias motoras izquierda vs derecha.

 Los filtros permiten que los métodos espaciales funcionen bien.

## 4. ¿Cómo se integran en un pipeline de MI?
Pipeline típico:

Preprocesamiento:

Filtro notch 50/60 Hz → IIR

Pasa bandas 8–30 Hz → FIR (frecuente)

Segmentación

Dividir en epochs antes/durante MI

Extracción de características:

Potencia en mu/beta

CSP (requiere buen filtrado)

Clasificación:

SVM, LDA, CNN, etc.
